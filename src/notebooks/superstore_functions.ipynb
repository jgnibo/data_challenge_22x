```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("superstore.csv", encoding='windows-1254')
display(df.head())

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df = df.sort_values(by="Order Date")
display(df.head())
```

```python
df_segment_grouped = df.groupby(df.Segment)
df_consumer = df_segment_grouped.get_group("Consumer")
print("consumer df head:")
display(df_consumer.head())

df_corporate = df_segment_grouped.get_group("Corporate")
print("corporate df head:")
display(df_corporate.head())

df_homeoff = df_segment_grouped.get_group("Home Office")
print("home office df head:")
display(df_homeoff.head())
```

```python
# Segmented aggregate sale/profit DataFrames - to be converted into quarterly info

consumer_agg_sales = pd.DataFrame({
    'Order Date': df_consumer['Order Date'],
    'Sales': df_consumer['Sales'],
    'Profit': df_consumer['Profit']
})
consumer_agg_sales = pd.pivot_table(
    data=consumer_agg_sales,
    index='Order Date',
    aggfunc=np.sum
)


corporate_agg_sales = pd.DataFrame({
    'Order Date': df_corporate['Order Date'],
    'Sales': df_corporate['Sales'],
    'Profit': df_corporate['Profit']
})
corporate_agg_sales = pd.pivot_table(
    data=corporate_agg_sales,
    index='Order Date',
    aggfunc=np.sum
)


home_agg_sales = pd.DataFrame({
    'Order Date': df_homeoff['Order Date'],
    'Sales': df_homeoff['Sales'],
    'Profit': df_homeoff['Profit']
})
home_agg_sales = pd.pivot_table(
    data=home_agg_sales,
    index='Order Date',
    aggfunc=np.sum
)

display(consumer_agg_sales.head())
display(corporate_agg_sales.head())
display(home_agg_sales.head())


# Segmented subcategory DataFrames, as a percentage of total sales

consumer_sub_sales = pd.DataFrame({
    'Subcategory': df_consumer['Sub-Category'],
    'Sales': df_consumer['Sales'],
    'Profit': df_consumer['Profit']
})
consumer_sub_sales = pd.pivot_table(
    data=consumer_sub_sales,
    index='Subcategory',
    aggfunc=np.sum
)
consumer_sub_sales['Sales Percentage'] = consumer_sub_sales['Sales'] / consumer_sub_sales['Sales'].sum() * 100
consumer_sub_sales['Profit Percentage'] = consumer_sub_sales['Profit'] / consumer_sub_sales['Profit'].sum() * 100


corporate_sub_sales = pd.DataFrame({
    'Subcategory': df_corporate['Sub-Category'],
    'Sales': df_corporate['Sales'],
    'Profit': df_corporate['Profit']
})
corporate_sub_sales = pd.pivot_table(
    data=corporate_sub_sales,
    index='Subcategory',
    aggfunc=np.sum
)
corporate_sub_sales['Sales Percentage'] = corporate_sub_sales['Sales'] / corporate_sub_sales['Sales'].sum() * 100
corporate_sub_sales['Profit Percentage'] = corporate_sub_sales['Profit'] / corporate_sub_sales['Profit'].sum() *100


home_sub_sales = pd.DataFrame({
    'Subcategory': df_homeoff['Sub-Category'],
    'Sales': df_homeoff['Sales'],
    'Profit': df_homeoff['Profit']
})
home_sub_sales = pd.pivot_table(
    data=home_sub_sales,
    index='Subcategory',
    aggfunc=np.sum
)
home_sub_sales['Sales Percentage'] = home_sub_sales['Sales'] / home_sub_sales['Sales'].sum() * 100
home_sub_sales['Profit Percentage'] = home_sub_sales['Profit'] / home_sub_sales['Profit'].sum() *100


print(consumer_sub_sales)
print(consumer_sub_sales['Sales Percentage'].sum())

print(consumer_sub_sales['Profit Percentage'].sum())

print(corporate_sub_sales)
print(corporate_sub_sales['Profit Percentage'].sum())
print(corporate_sub_sales['Sales Percentage'].sum())


print(home_sub_sales)
print(home_sub_sales['Profit Percentage'].sum())
print(home_sub_sales['Sales Percentage'].sum())
```

```python
def detect_outliers_zscore(data, threshhold):
    outliers_index = []
    outliers = []
    mean = np.mean(data)
    std = np.std(data)
    for index, value in enumerate(data):
        z_score = (value-mean)/std
        if(np.abs(z_score) > threshhold):
            outliers_index.append(index)
    
    return outliers_index

print(detect_outliers_zscore(consumer_agg_sales['Sales'], 3))
```

```python
def replace_outliers(data, threshhold):
    outliers_index = detect_outliers_zscore(data, threshhold)
    median = np.median(data)
    for i in outliers_index:
         data[i] = median
```

```python
replace_outliers(consumer_agg_sales['Profit'], 3)
replace_outliers(consumer_agg_sales['Sales'], 3)
replace_outliers(corporate_agg_sales['Profit'], 3)
replace_outliers(corporate_agg_sales['Sales'], 3)
replace_outliers(home_agg_sales['Profit'], 3)
replace_outliers(home_agg_sales['Sales'], 3)
```
