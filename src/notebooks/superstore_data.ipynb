```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("superstore.csv", encoding='windows-1254')
display(df.head())

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df = df.sort_values(by="Order Date")
display(df.head())
```

```python
df_segment_grouped = df.groupby(df.Segment)
df_consumer = df_segment_grouped.get_group("Consumer")
print("consumer df head:")
display(df_consumer.head())

df_corporate = df_segment_grouped.get_group("Corporate")
print("corporate df head:")
display(df_corporate.head())

df_homeoff = df_segment_grouped.get_group("Home Office")
print("home office df head:")
display(df_homeoff.head())
```

```python
print("describe consumer df")
consumer_describe = df_consumer.describe()
display(consumer_describe)

print("describe corporate df")
corporate_describe = df_corporate.describe()
display(corporate_describe)

print("describe home office df")
display(df_homeoff.describe())
```

```python
print(type(df_consumer['Order Date'][0]))
display(df_consumer.head())
```

```python
df_con_sum = pd.DataFrame(index=pd.date_range(df_consumer["Order Date"].min(), df_consumer["Order Date"].max()), columns = ['Sales'])
df_con_sum['Sales'] = df_con_sum.apply(lambda x: df_consumer.loc[(df['Order Date'] <= x.name), ['Sales']].sum(), axis=1)

df_cor_sum = pd.DataFrame(index=pd.date_range(df_corporate["Order Date"].min(), df_corporate["Order Date"].max()), columns = ['Sales'])
df_cor_sum['Sales'] = df_cor_sum.apply(lambda x: df_corporate.loc[(df['Order Date'] <= x.name), ['Sales']].sum(), axis=1)

df_home_sum = pd.DataFrame(index=pd.date_range(df_homeoff["Order Date"].min(), df_homeoff["Order Date"].max()), columns = ['Sales'])
df_home_sum['Sales'] = df_home_sum.apply(lambda x: df_homeoff.loc[(df['Order Date'] <= x.name), ['Sales']].sum(), axis=1)






```

```python
plt.plot(df_con_sum)
plt.plot(df_cor_sum)
plt.plot(df_home_sum)
```
